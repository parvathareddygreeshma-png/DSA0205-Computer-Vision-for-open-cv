```
# Install OpenCV (if needed)
!pip install opencv-python

# Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read image
image = cv2.imread(filename)

# Check image
if image is None:
    print("Error: Unable to read the image.")
else:
    # Convert image from BGR to LAB color space
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

    # Split LAB channels
    l, a, b = cv2.split(lab)

    # Apply CLAHE for image enhancement
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    l_enhanced = clahe.apply(l)

    # Merge channels
    enhanced_lab = cv2.merge((l_enhanced, a, b))

    # Convert back to BGR
    enhanced_bgr = cv2.cvtColor(enhanced_lab, cv2.COLOR_LAB2BGR)

    # Convert to RGB for displaying
    original_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    enhanced_rgb = cv2.cvtColor(enhanced_bgr, cv2.COLOR_BGR2RGB)

    # Save enhanced image
    cv2.imwrite("enhanced_image.jpg", enhanced_bgr)

    # Display images
    plt.figure(figsize=(12,6))

    plt.subplot(1,2,1)
    plt.imshow(original_rgb)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(enhanced_rgb)
    plt.title("Enhanced Image")
    plt.axis("off")

    plt.show()

    print("Enhanced image saved as 'enhanced_image.jpg'")
```

```
# Install OpenCV
!pip install opencv-python

# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read image in grayscale
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Check image
if image is None:
    print("Error: Unable to read the image.")
else:
    # Convert image to float
    image_float = image.astype(np.float32)

    # Apply Log Transformation
    c = 255 / np.log(1 + np.max(image_float))
    log_transformed = c * np.log(1 + image_float)

    # Convert to unsigned 8-bit integer
    log_transformed = np.array(log_transformed, dtype=np.uint8)

    # Display images
    plt.figure(figsize=(12,6))

    plt.subplot(1,2,1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(log_transformed, cmap='gray')
    plt.title("Log Transformed Image")
    plt.axis("off")

    plt.show()

    # Save output
    cv2.imwrite("log_transformed_image.jpg", log_transformed)
    print("Log transformed image saved as 'log_transformed_image.jpg'")
```

```
# Install OpenCV
!pip install opencv-python

# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read image in grayscale
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Check image
if image is None:
    print("Error: Unable to read the image.")
else:
    # Normalize image (0 to 1)
    image_norm = image / 255.0

    # Gamma value (change this value as needed)
    gamma = 0.5

    # Apply Power Law (Gamma) Transformation
    power_law = np.power(image_norm, gamma)

    # Scale back to 0-255
    power_law = np.uint8(power_law * 255)

    # Display images
    plt.figure(figsize=(12,6))

    plt.subplot(1,2,1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(power_law, cmap='gray')
    plt.title("Power Law Transformed Image")
    plt.axis("off")

    plt.show()

    # Save output
    cv2.imwrite("power_law_transformed.jpg", power_law)
    print("Power Law transformed image saved as 'power_law_transformed.jpg'")
```

```
# Install OpenCV
!pip install opencv-python

# Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read image in grayscale
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Check image
if image is None:
    print("Error: Unable to read the image.")
else:
    # Apply Histogram Equalization
    equalized = cv2.equalizeHist(image)

    # Display Original and Equalized Images
    plt.figure(figsize=(12,6))

    plt.subplot(1,2,1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(equalized, cmap='gray')
    plt.title("Histogram Equalized Image")
    plt.axis("off")

    plt.show()

    # Plot Histograms
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.hist(image.ravel(), bins=256, range=[0,256], color='blue')
    plt.title("Histogram of Original Image")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

    plt.subplot(1,2,2)
    plt.hist(equalized.ravel(), bins=256, range=[0,256], color='green')
    plt.title("Histogram of Equalized Image")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

    plt.show()

    # Save output
    cv2.imwrite("equalized_image.jpg", equalized)
    print("Histogram equalized image saved as 'equalized_image.jpg'")
```

```
# Install required libraries
!pip install opencv-python scikit-image

# Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from skimage.exposure import match_histograms

# Upload Source Image
print("Upload Source Image")
uploaded = files.upload()
source_file = list(uploaded.keys())[0]

# Upload Reference Image
print("Upload Reference Image")
uploaded = files.upload()
reference_file = list(uploaded.keys())[0]

# Read images in grayscale
source = cv2.imread(source_file, cv2.IMREAD_GRAYSCALE)
reference = cv2.imread(reference_file, cv2.IMREAD_GRAYSCALE)

# Check images
if source is None or reference is None:
    print("Error: Unable to read one or both images.")
else:
    # Histogram Specification (Matching)
    matched = match_histograms(source, reference)

    # Convert to uint8
    matched = matched.astype('uint8')

    # Display images
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(source, cmap='gray')
    plt.title("Source Image")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(reference, cmap='gray')
    plt.title("Reference Image")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(matched, cmap='gray')
    plt.title("Histogram Matched Image")
    plt.axis("off")

    plt.show()

    # Plot Histograms
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.hist(source.ravel(), 256, [0,256], color='blue')
    plt.title("Source Histogram")

    plt.subplot(1,3,2)
    plt.hist(reference.ravel(), 256, [0,256], color='green')
    plt.title("Reference Histogram")

    plt.subplot(1,3,3)
    plt.hist(matched.ravel(), 256, [0,256], color='red')
    plt.title("Matched Histogram")

    plt.show()

    # Save output
    cv2.imwrite("histogram_specification.jpg", matched)
    print("Histogram specification completed successfully.")
```
